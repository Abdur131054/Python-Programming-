**Definition:**
Modules are code libraries in Python that contain functions, classes, and variables we can reuse in your programs.

**Purpose:**
They help organize code and avoid repetition â€” we can import them instead of writing everything from scratch.

| Module     | Purpose                            | Example Use                           |
| ---------- | ---------------------------------- | ------------------------------------- |
| **math**   | Mathematical operations            | `math.sqrt(16)` â†’ `4.0`               |

| **os**     | Interact with the operating system | `os.getcwd()` â†’ current directory     |
| **time**   | Handle time-related tasks          | `time.sleep(2)` â†’ pause for 2 seconds |
